# Q1. Kubernetes Setup: Set up a local Kubernetes cluster using      Minikube or a
managed Kubernetes service ? (e.g., GKE, EKS).

- To set up a local Kubernetes cluster, **Minikube** is one of the easiest options. Here is a step-by-step guide.

# Kubernetes Setup Using Minikube

## Step 1: Install Docker

Download and install Docker Desktop (or Docker Engine on Linux), then verify the installation:

```bash
docker --version
```

## Step 2: Install kubectl

### On Windows (using Chocolatey)

```bash
choco install kubernetes-cli
```

### On Ubuntu

```bash
sudo apt update
sudo apt install -y kubectl
```

Verify:

```bash
kubectl version --client
```

## Step 3: Install Minikube

### Windows (Chocolatey)

```bash
choco install minikube
```

### Ubuntu

```bash
curl -LO https://storage.googleapis.com/minikube/releases/latest/minikube-linux-amd64
sudo install minikube-linux-amd64 /usr/local/bin/minikube
```

Verify:

```bash
minikube version
```

## Step 4: Start the Kubernetes Cluster

```bash
minikube start --driver=docker
```

This command creates a single-node local Kubernetes cluster.

## Step 5: Check Cluster Status

```bash
minikube status
```

Expected output:

```
host: Running
kubelet: Running
apiserver: Running
kubeconfig: Configured
```

## Step 6: Verify Kubernetes Nodes

```bash
kubectl get nodes
```

Example output:

```
NAME       STATUS   ROLES           AGE   VERSION
minikube   Ready    control-plane   1m    v1.xx.x
```

## Step 7: Deploy a Sample Application

Create a deployment:

```bash
kubectl create deployment nginx --image=nginx
```

Expose it as a service:

```bash
kubectl expose deployment nginx --type=NodePort --port=80
```

Check resources:

```bash
kubectl get deployments
kubectl get pods
kubectl get services
```

## Step 8: Access the Application

Run:

```bash
minikube service nginx
```

This opens the NGINX page in your browser.

## Step 9: Stop or Delete the Cluster

Stop:

```bash
minikube stop
```

Delete:

```bash
minikube delete
```

---

## Alternative: Managed Kubernetes Services

If you prefer a cloud-hosted cluster instead of Minikube, you can use:

* Google Cloud **Google Kubernetes Engine (GKE)** – Managed Kubernetes on Google Cloud.
* Amazon Web Services **Amazon Elastic Kubernetes Service (EKS)** – Managed Kubernetes on AWS.
* Microsoft **Azure Kubernetes Service (AKS)** – Managed Kubernetes on Azure.

These services handle the control plane for you and are commonly used in production environments.




# Q2. Application Deployment: Deploy a simple application to your Kubernetes
cluster?

- # Deploy a Simple Application to a Kubernetes Cluster

This example deploys an **NGINX** web server to a Kubernetes cluster (such as Minikube).

## 1. Create a Deployment

Run the following command:

```bash
kubectl create deployment nginx-app --image=nginx:latest
```

Verify that the deployment was created:

```bash
kubectl get deployments
kubectl get pods
```

Expected output:

```text
NAME         READY   UP-TO-DATE   AVAILABLE   AGE
nginx-app    1/1     1            1           30s
```

---

## 2. Expose the Deployment as a Service

Create a `NodePort` service so the application can be accessed:

```bash
kubectl expose deployment nginx-app \
  --type=NodePort \
  --port=80 \
  --target-port=80
```

Check the service:

```bash
kubectl get services
```

Example output:

```text
NAME         TYPE       CLUSTER-IP      EXTERNAL-IP   PORT(S)        AGE
nginx-app    NodePort   10.96.120.100   <none>        80:31234/TCP   10s
```

---

## 3. Access the Application

If you're using **Minikube**, run:

```bash
minikube service nginx-app
```

This will open the application in your default web browser.

Alternatively, get the Minikube IP:

```bash
minikube ip
```

Then access:

```text
http://<MINIKUBE_IP>:<NODE_PORT>
```

Replace `<MINIKUBE_IP>` and `<NODE_PORT>` with the values from your environment.

---

## 4. Deploy Using a YAML Manifest (Recommended)

Create a file named `nginx-deployment.yaml`:

```yaml
apiVersion: apps/v1
kind: Deployment
metadata:
  name: nginx-app
spec:
  replicas: 2
  selector:
    matchLabels:
      app: nginx
  template:
    metadata:
      labels:
        app: nginx
    spec:
      containers:
      - name: nginx
        image: nginx:latest
        ports:
        - containerPort: 80
---
apiVersion: v1
kind: Service
metadata:
  name: nginx-service
spec:
  type: NodePort
  selector:
    app: nginx
  ports:
  - port: 80
    targetPort: 80
```

Apply it:

```bash
kubectl apply -f nginx-deployment.yaml
```

Verify:

```bash
kubectl get all
```

---

## 5. Scale the Application

Increase the number of replicas:

```bash
kubectl scale deployment nginx-app --replicas=3
```

Check the updated pods:

```bash
kubectl get pods
```

---

## 6. Clean Up

Delete the deployment and service:

```bash
kubectl delete deployment nginx-app
kubectl delete service nginx-app
# or, if you used the YAML above:
kubectl delete -f nginx-deployment.yaml
```

Pods, Services, and Deployments?

- # Kubernetes Resource Management Practice (Pods, Services, and Deployments)

This exercise demonstrates how to create, inspect, update, scale, and delete the most common Kubernetes resources.

## 1. Manage Pods

### Create a Pod

```bash
kubectl run demo-pod --image=nginx
```

### View Pods

```bash
kubectl get pods
```

### Get Detailed Information

```bash
kubectl describe pod demo-pod
```

### View Pod Logs

```bash
kubectl logs demo-pod
```

### Delete the Pod

```bash
kubectl delete pod demo-pod
```

---

## 2. Manage Deployments

### Create a Deployment

```bash
kubectl create deployment web-app --image=nginx
```

### View Deployments

```bash
kubectl get deployments
```

### Scale the Deployment to 3 Replicas

```bash
kubectl scale deployment web-app --replicas=3
```

Verify:

```bash
kubectl get pods
```

### Update the Container Image

```bash
kubectl set image deployment/web-app nginx=nginx:1.27
```

### Check Rollout Status

```bash
kubectl rollout status deployment/web-app
```

### Roll Back to the Previous Version (if needed)

```bash
kubectl rollout undo deployment/web-app
```

### Delete the Deployment

```bash
kubectl delete deployment web-app
```

---

## 3. Manage Services

### Expose the Deployment

```bash
kubectl expose deployment web-app \
  --type=NodePort \
  --port=80 \
  --target-port=80
```

### List Services

```bash
kubectl get services
```

### Describe the Service

```bash
kubectl describe service web-app
```

### Access the Service (Minikube)

```bash
minikube service web-app
```

### Delete the Service

```bash
kubectl delete service web-app
```

---

## 4. Check All Resources

Display the main resources in the current namespace:

```bash
kubectl get all
```

Example output:

```text
NAME                            READY   STATUS
pod/web-app-7c9d6d5b8f-abc12    1/1     Running
pod/web-app-7c9d6d5b8f-def34    1/1     Running
pod/web-app-7c9d6d5b8f-ghi56    1/1     Running

NAME                 TYPE       CLUSTER-IP      PORT(S)
service/web-app      NodePort   10.96.10.120    80:31234/TCP

NAME                      READY   UP-TO-DATE   AVAILABLE
deployment.apps/web-app   3/3     3            3
```

---

## 5. Practice Using YAML

Create a file named `web-app.yaml`:

```yaml
apiVersion: apps/v1
kind: Deployment
metadata:
  name: web-app
spec:
  replicas: 2
  selector:
    matchLabels:
      app: web-app
  template:
    metadata:
      labels:
        app: web-app
    spec:
      containers:
        - name: nginx
          image: nginx:latest
          ports:
            - containerPort: 80
---
apiVersion: v1
kind: Service
metadata:
  name: web-app-service
spec:
  selector:
    app: web-app
  ports:
    - port: 80
      targetPort: 80
  type: NodePort
```

Apply it:

```bash
kubectl apply -f web-app.yaml
```

Delete it:

```bash
kubectl delete -f web-app.yaml
```

Kubernetes?

- # Using Helm to Package and Deploy Applications on Kubernetes

**Helm** is a package manager for Kubernetes. It simplifies deploying and managing applications by using reusable templates called **Helm Charts**.

## Step 1: Install Helm

Verify Helm is installed:

```bash
helm version
```

If it is not installed:

* **Windows (Chocolatey):**

  ```bash
  choco install kubernetes-helm
  ```

* **Ubuntu/Linux:**

  ```bash
  curl https://raw.githubusercontent.com/helm/helm/main/scripts/get-helm-3 | bash
  ```

---

## Step 2: Create a New Helm Chart

Generate a sample chart:

```bash
helm create my-nginx
```

This creates a directory structure like:

```text
my-nginx/
├── Chart.yaml
├── values.yaml
├── charts/
└── templates/
    ├── deployment.yaml
    ├── service.yaml
    ├── ingress.yaml
    └── ...
```

---

## Step 3: Customize the Chart

Edit `values.yaml` to set the image and replica count:

```yaml
replicaCount: 2

image:
  repository: nginx
  tag: latest
```

You can also modify files in the `templates/` directory if needed.

---

## Step 4: Deploy the Chart

Install the chart into your Kubernetes cluster:

```bash
helm install my-release ./my-nginx
```

Verify the release:

```bash
helm list
```

Check the Kubernetes resources:

```bash
kubectl get all
```

---

## Step 5: Upgrade the Application

After changing `values.yaml` (for example, increasing replicas to 3):

```yaml
replicaCount: 3
```

Apply the update:

```bash
helm upgrade my-release ./my-nginx
```

Confirm the change:

```bash
kubectl get deployments
```

---

## Step 6: View Release Details

```bash
helm status my-release
```

See all installed releases:

```bash
helm list
```

View release history:

```bash
helm history my-release
```

---

## Step 7: Roll Back to a Previous Revision

If an upgrade causes issues:

```bash
helm rollback my-release 1
```

Replace `1` with the desired revision number from `helm history`.

---

## Step 8: Uninstall the Application

Remove the Helm release and its associated Kubernetes resources:

```bash
helm uninstall my-release
```

---

## Example: Deploy an Existing Helm Chart

You can also install a chart from a Helm repository. For example:

```bash
helm repo add bitnami https://charts.bitnami.com/bitnami
helm repo update
helm install my-webserver bitnami/nginx
```

---
